```python
#symbols = sp.symbols('x1 x2')
x1, x2 = sp.symbols('x1 x2')
f_A = -2*x1**2 + 3*x1*x2 - 1.5*x2**2 - 1.3
x_init=np.array([-4,4])
x_best, fx_best, iters = hill_climber(f_A, delta=0.1, n_iter=1000, x_init=x_init, constrain = [[-6, 6],[-6, 6]])
print(f'min f_B(x1,x2) = f({x_best[0]:.2f}, {x_best[1]:.2f}) = {fx_best:.2f} , Iterations: {iters}')
```

```python
#symbols = sp.symbols('x1 x2')
x1, x2 = sp.symbols('x1 x2')
f_A = -2*x1**2 + 3*x1*x2 - 1.5*x2**2 - 1.3
x_init=np.array([-4,4])
x_best, fx_best, iters = grad_descent(f_A, tol=.001, x_init=x_init, max_iter = 1000, constrain = [[-6, 6],[-6, 6]])
print(f'min f_B(x1,x2) = f({x_best[0]:.2f}, {x_best[1]:.2f}) = {fx_best:.2f} , Iterations: {iters}')
```
